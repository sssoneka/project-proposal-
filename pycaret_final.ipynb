```python
import pandas as pd
from pycaret.regression import *
data = pd.read_excel('END.xlsx')
```

```python
exp = setup(
    data,
    target='Points',
    ignore_features=['lag_Popularity',
                     'lag_Popularity_sq',
                     'Popularity',
                     'Popularity_sq',
                     'lag_Points',
                     'Name', 'Rank', 'Time'],
)

best_model = compare_models(include=['et', 'rf', 'dt', 'gbr', 'lr'])
results = pull()
res = results[['MAE', 'R2']]
print(res)
```
